#### Univariate Analysis
*Univariate analysis is simplest form of analysis.
*Only one variable is used in visualisation analysis.This does not deal with relationship plots.
*It takes data, summarize the data, pattern analysis, trend analysis in data. 
*Univariate analysis can be implemented on **numerical data** or **categorical data**

```python
df_numerical=housing_data.select_dtypes(include=['number'])
df_numerical
```

```python
df_categorical=housing_data.select_dtypes(include=['object'])
df_categorical
```

#### 1.Line Plot
*Univariate data-Trends Analysis
*Bivariate data-Relationship Plot

```python
#Plot using dataframe and plottype
df_numerical['SalePrice'].plot() #kind parameter can be set to different types of plots like piechart,area,line,bar etc
#it is inbuilt function to pandas to create plot
```

```python
sorted_data=df_numerical['SalePrice'].sort_values(ascending=True).unique()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,8))
plt.plot(sorted_data,color='orangered')
plt.xlabel('Index')
plt.ylabel('Price')
#size= set the front size for Title and Labels
plt.title('Trends of Sale Price', size=20)
plt.show()
```

#### Bar Plot
*Univariate or Bivariate Plots
*Bar chart are used with categorical dataset
*Bar plots are used to display counts of unique value for categorical type, height of bar indicate count for each categorical

```python
df_categorical['SaleType'].value_counts()
```

```python
print(df_categorical['SaleType'].value_counts().values)
print(df_categorical['SaleType'].value_counts().index)

```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,8))
plt.bar(df_categorical['SaleType'].value_counts().index,color='orangered',height=df_categorical['SaleType'].value_counts().values)
plt.xlabel('SaleType')
plt.ylabel('Count')
#size= set the front size for Title and Labels
plt.title('Trends of Sale Type', size=20)
plt.show()
```

```python
df_categorical['SaleType'].value_counts().plot(kind='bar',figsize=(11,7),ylabel='Count')
```

#### Histogram
*Univariate Analysis- Apply on continuos Data
*Frequency of data in range.
*Study of Spread/ data distribution

```python
#bins= 20, Indicates 20 equal distribution of Numerical Data
sns.set_style('whitegrid')
plt.hist(df_numerical['SalePrice'],bins=20)
plt.ylabel('Count')
plt.xlabel('Price')
#size= set the front size for Title and Labels
plt.title('Sale Price Histogram', size=20)
plt.show()
```

```python
df_numerical['SalePrice'].plot(kind='hist',bins=20,xlabel="Sale Price",ylabel='Count',title="Price Histogram",color='Blue')
```

#### Scatter Plot
*Indicates relationship Plot or Correlation between two numerical columns

```python
df_numerical
```

```python
#Considering points with lotarea<60000 as data is concentrated in that region
plt.scatter(df_numerical[df_numerical['LotArea']<60000]['LotArea'], 
            df_numerical[df_numerical['LotArea']<60000]['SalePrice'])

plt.xlabel("LotArea")
plt.ylabel("SalePrice")
plt.title("Relationship b/w LotArea And SalePrice")
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=df_numerical[df_numerical['LotArea']<60000]['LotArea'], 
            y=df_numerical[df_numerical['LotArea']<60000]['SalePrice'],hue=df_categorical['HouseStyle'],s=200)
plt.show() #Compulsory for the plot to appear for anywhere other than jupyter notebook
```

```python
plt.figure(figsize=(10,8))

#s=size of bubble, marker='s','^','>','<','*','.','+','d','o'
sns.scatterplot(x=df_numerical['GarageArea'], 
            y=df_numerical['SalePrice'],hue=df_categorical['HouseStyle'],s=200,alpha=0.5,marker='*')
plt.show() #Compulsory for the plot to appear for anywhere other than jupyter notebook
```
