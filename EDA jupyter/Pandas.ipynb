#### 11. set and reset index

```python
import pandas as pd

# Example data: City names and their per capita GDP (in USD)
data = {
    'City': ['New York', 'Berlin', 'Tokyo', 'São Paulo', 'Mumbai'],
    'Per Capita GDP (USD)': [85000, 60000, 50000, 15000, 3000]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set 'City' as the index
df.set_index('City', inplace=True)

# Convert DataFrame to Series
gdp_series = df['Per Capita GDP (USD)']

print(gdp_series)
```

```python
city_gdp=pd.DataFrame(gdp_series)
```

```python
city_gdp
```

```python
city_gdp=city_gdp.reset_index()
city_gdp
```

```python
city_gdp=city_gdp.set_index('City')
city_gdp
```

```python
city_gdp=city_gdp.reset_index('City')
city_gdp
```

#### 13.Merge

```python
import pandas as pd

# DataFrame 1: City and their population
data1 = {
    'City': ['New York', 'Berlin', 'Tokyo', 'São Paulo', 'Mumbai'],
    'Population': [8419000, 3645000, 13960000, 12300000, 12480000]
}
df1 = pd.DataFrame(data1)

# DataFrame 2: City and their per capita GDP
data2 = {
    'City': ['New York', 'Berlin', 'Tokyo', 'São Paulo', 'Mumbai'],
    'gdp_per_capita': [85000, 60000, 50000, 15000, 3000]
}
df2 = pd.DataFrame(data2)

print("DataFrame 1:")
print(df1)

print("\nDataFrame 2:")
print(df2)
```

```python
pd.merge(df1,df2,on='City',how='inner')
```

```python
#to rename column names:
#df1.columns=['city','population','gdp']
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```
