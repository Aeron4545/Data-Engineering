#### 2.Restaurant with Online Order

```python
y=len((zomato_df[zomato_df['online_order']=='Yes']))/(zomato_df['online_order'].count())
n=len((zomato_df[zomato_df['online_order']=='No']))/(zomato_df['online_order'].count())
#p=pd.DataFrame(['Yes','No'],[y,n])
p=pd.DataFrame({'Online Order':['Yes','No'],'Percentage':[y*100,n*100]})
print(p)
plt.pie([y,n],labels=['Yes','No'],autopct='%0.2f%%',colors=['red','mistyrose'])
plt.show()
```

```python
#alternate method
p=zomato_df[['online_order']].value_counts()/len(zomato_df)
print(p)
plt.pie(p,labels=['Yes','No'],autopct='%0.2f%%',colors=['red','mistyrose'])
plt.show()
```

```python
#alternate method
p=zomato_df['online_order'].value_counts()/zomato_df['online_order'].shape[0]
print(p)
plt.pie(p,labels=['Yes','No'],autopct='%0.2f%%',colors=['red','mistyrose'])
plt.show()
```

```python
locations=pd.DataFrame({'Name':zomato_df['location'].unique()})
locations['Name']=locations['Name'].apply(lambda x: "Banglore "+str(x))
geolocator=Nominatim(user_agent="app")

lat_long=[]

for location in locations['Name']:
    location=geolocator.geocode(location)
    if location is None:
        lat_long.append(np.nan)
    else:
        geo=(location.latitude,location.longitude)
        lat_long.append(geo)
        lat_long.append(geo)
        
```
